Давайте поанализируем наш алгоритм
Поймём что наш алгоритм всегда работает за нужную асимптотику, соответственно на время проверять его не имеет значение также как и на память
Будем проверять его на точность, насколько он близок к нашему правильному ответу

Так как задача для нахождения истинного ответа трудная и долгая, будем делать маленькие графы до 20 вершин, чтобы перебор работал не слишком долго
Запустим на 2 выборках, в одной из них у нас не плотный граф, в других плотный, посмотреть как я генерю, можно в коде(за это отвечает флаг)

Рассмотрим когда граф не плотный

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', font_scale=1.3, palette='Set2')
%matplotlib inline
```

```python
myfile = open("test.out", "r")
myline = myfile.readline()
n = int(myline)
our = []
perfect = []
diff = []
diffproc = []

for i in range(n):
    myline = myfile.readline()
    our.append(int(myline))
    myline = myfile.readline()
    perfect.append(int(myline))
    myline = myfile.readline()
    diff.append(int(myline))
    myline = myfile.readline()
    diffproc.append(float(myline))
myfile.close()  
```

```python
len(diffproc)
```

```python
np.mean(diffproc)
```

Размер выборки

```python
diffproc.sort()
plt.title('Гистограмма насколько точно приближен ответ')
sns.histplot(data=diffproc)
plt.xlabel('во сколько раз больше наш ответ')
plt.ylabel('количество');
```

из графика видно, что на не плотных графах даёт приблизительно хороший ответ, около половины даёт такой же ответ как и оптимальный

Теперь рассмотрим плотные графы

```python
myfile = open("test.out", "r")
myline = myfile.readline()
n = int(myline)
our = []
perfect = []
diff = []
diffproc_plot = []

for i in range(n):
    myline = myfile.readline()
    our.append(int(myline))
    myline = myfile.readline()
    perfect.append(int(myline))
    myline = myfile.readline()
    diff.append(int(myline))
    myline = myfile.readline()
    diffproc_plot.append(float(myline))
myfile.close()  
len(diffproc_plot)
```

```python
diffproc.sort()
plt.title('Гистограмма насколько точно приближен ответ')
sns.histplot(data=diffproc_plot)
plt.xlabel('во сколько раз больше наш ответ')
plt.ylabel('количество');
```

Видим, что на плотных результаты похуже, но тоже около половина графов даёт оптимальный результат

Давайте посчитаем сколько он в среднем

```python
np.mean(diffproc_plot)
```

Из всего анализа можно увидеть, что алгоритм даёт хорошее приближение, в среднем ошибается на 8 процентов, и в половине случаев даёт оптимальный ответ. В плотных графах работает похуже, что в принципе логично, так как они сложнее устроены,
